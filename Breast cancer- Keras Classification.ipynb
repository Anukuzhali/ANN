```python
X=df.drop('benign_0__mal_1',axis=1).values
```

```python
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation
```

```python
model=Sequential()

model.add(Dense(units=30,activation='relu'))

model.add(Dense(units=15,activation='relu'))

model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=500,
          validation_data=(X_test, y_test), verbose=1
          )
```

```python
model_loss=pd.DataFrame(model.history.history)
```

```python
model_loss
```

```python
model_loss.plot()
```

## Early Stopping

We obviously trained too much! Let's use early stopping to track the val_loss and stop training once it begins increasing too much!

```python
model1=Sequential()

model1.add(Dense(units=30,activation='relu'))

model1.add(Dense(units=15,activation='relu'))

model1.add(Dense(units=1,activation='sigmoid'))

model1.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model1.fit(x=X_train,
          y=y_train,
          epochs=500,
          validation_data=(X_test,y_test),verbose=1,
          callbacks=[early_stop])
```

```python
model1_loss=pd.DataFrame(model1.history.history)
```

```python
model1_loss
```

```python
model1_loss.plot()
```

## Adding in Dropout Layers

```python
from tensorflow.keras.layers import Dropout
```

```python
model2=Sequential()

model2.add(Dense(units=30,activation='relu'))
model2.add(Dropout(0.5))

model2.add(Dense(units=15,activation='relu'))
model2.add(Dropout(0.5))

model2.add(Dense(units=1,activation='sigmoid'))

model2.compile(loss='binary_crossentropy',optimizer='adam')

```

```python
model2.fit(x=X_train,
          y=y_train,
          epochs=500,
          validation_data=(X_test,y_test),verbose=1,
          callbacks=[early_stop])
           
```

```python
model2_loss=pd.DataFrame(model2.history.history)
```

```python
model2_loss.plot()
```

```python
predictions=model2.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
confusion_matrix(y_test,predictions)
```
